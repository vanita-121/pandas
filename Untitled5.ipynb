```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder,OneHotEncoder
```

```python
df=pd.read_csv("encoder2(customer_with_colors) (1).csv")
df
```

```python
a = df.iloc[:,1:3]
le = OrdinalEncoder(categories=[["Poor","Average","Good"],["School","UG","PG"]])
aa = le.fit_transform(a)
print(aa)
```

```python
# label encoder
lle = LabelEncoder()
lle.fit_transform(df["purchased"])
```

```python
y=df.iloc[:,-1].values.reshape(-1,1)
ohe=OneHotEncoder()
y=ohe.fit_transform(y).toarray()
print(y)
```
